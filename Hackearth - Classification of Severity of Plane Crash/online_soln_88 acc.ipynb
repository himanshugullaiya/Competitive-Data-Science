### Selecting Best Features

```python
x = dataset.iloc[:,1:]
y = dataset.iloc[:,0]
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x,y)
model.feature_importances_
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.sort_values(ascending=True).plot(kind='barh',cmap='viridis')
plt.show()
```

### Plotting Graphs of Most relevant Features

```python
from scipy.stats import norm
```

```python
cols = ['Safety_Score', 'Days_Since_Inspection', 'Control_Metric', 'Accident_Type_Code']
plt.title('Distribution Plot')
for col in cols:    
    chart = sns.distplot(dataset[col], fit = norm)
    (mu, sigma) = norm.fit(dataset[col])
    chart.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],loc='best')
    plt.ylabel('Frequency')
    plt.show()
```

### Feature Plot

```python
best_feat = ['Safety_Score','Days_Since_Inspection','Control_Metric','Accident_Type_Code']
for feature in best_feat[1:]:
    sns.countplot(dataset[feature])
    plt.show()
```

### Splitting Dataset

```python
x = dataset[best_feat]
y = dataset['Severity']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```
